```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
pd.set_option('display.max_columns', None)
```

```python
df_matches = pd.read_csv('2022_to_2023_all_matches.csv')
```

```python
#Engineer the Run Rate according to the following rules 
def getRequiredRunRate_first_innings(over , wickets):
  if over < 6:
    if wickets < 1:
      return 10
    elif wickets>=1 and wickets<4:
      return  8
    else:
      6
  elif over >= 6 and over<15:
    if wickets < 1:
      return 11
    elif wickets>=1 and wickets<4:
      return  9
    elif wickets>=4 and wickets<6:
      return 8
    else:
      return 6  
  elif over >= 15 and over<20:
    if wickets < 1:
      return 15
    elif wickets>=1 and wickets<7:
      return  12
    else :
      return 10
    # code block 3

current_index = 0

while current_index < len(df_matches):
   if(df_matches['innings'][current_index] == 1):
     #print('Addding the First Innings Required Run Rate index is ' ,current_index)
     df_matches['req_run_rate'][current_index] = getRequiredRunRate_first_innings(df_matches['over'][current_index] , df_matches['over_start_total_wickets_fallen'][current_index])
   current_index = current_index + 1
```
